```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample retail sales dataset
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Sales': [50000, 35000, 20000, 10000, 15000],
    'Profit': [12000, 8000, 5000, 2500, 4000],
    'Region': ['North', 'South', 'East', 'West', 'North']
}

df = pd.DataFrame(data)

# Display dataset
print("Retail Dataset:")
print(df)

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Total sales and profit
print("\nTotal Sales:", df['Sales'].sum())
print("Total Profit:", df['Profit'].sum())

# Visualization 1: Sales by Product
plt.figure(figsize=(6,4))
plt.bar(df['Product'], df['Sales'])
plt.title('Sales by Product')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# Visualization 2: Profit by Product
plt.figure(figsize=(6,4))
plt.bar(df['Product'], df['Profit'])
plt.title('Profit by Product')
plt.xlabel('Product')
plt.ylabel('Profit')
plt.show()

# Visualization 3: Sales Distribution
plt.figure(figsize=(6,4))
plt.pie(df['Sales'], labels=df['Product'], autopct='%1.1f%%')
plt.title('Sales Distribution')
plt.show()

# Save dataset
df.to_csv('retail_sales_data.csv', index=False)

print("\nReal-World Retail Data Project Completed Successfully!")
```
